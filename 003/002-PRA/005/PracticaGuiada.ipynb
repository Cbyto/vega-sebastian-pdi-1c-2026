```python
from pathlib import Path

import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage as ski


def mostrar_imagenes(imagenes, titulos, cmap=None, figsize=(14, 4)):
    fig, axes = plt.subplots(1, len(imagenes), figsize=figsize)
    if len(imagenes) == 1:
        axes = [axes]

    for ax, imagen, titulo in zip(axes, imagenes, titulos):
        if imagen.ndim == 2:
            ax.imshow(imagen, cmap=cmap or "gray")
        else:
            ax.imshow(imagen)
        ax.set_title(titulo)
        ax.axis("off")

    plt.tight_layout()
    plt.show()


def segmentar_por_umbral(imagen, umbral):
    mascara = np.zeros_like(imagen, dtype=np.uint8)
    mascara[imagen > umbral] = 255
    return mascara
```

```python
base = Path(".")

cv2.imwrite(str(base / "paisaje.png"), cv2.cvtColor(ski.data.astronaut(), cv2.COLOR_RGB2BGR))
cv2.imwrite(str(base / "texto.png"), ski.data.page())
cv2.imwrite(str(base / "monedas.png"), ski.data.coins())

print("Archivos de práctica listos: paisaje.png, texto.png y monedas.png")
```

## Ejercicio 2: Recorte y redimensionado

1. Definí una región de interés de `100 x 100` píxeles dentro de `paisaje.png`.
2. Redimensioná ese recorte a `220 x 220` píxeles.
3. Armá un mosaico `2 x 2` con el recorte redimensionado.
4. Explicá, con una frase, la diferencia entre recortar y redimensionar.


```python
# Definí los índices de tu región de interés.
y1, y2 = 50, 200
x1, x2 = 100, 300

roi = img_rgb[y1:y2, x1:x2]
roi_grande = cv2.resize(roi, (220, 220))

# Redimensiono la imagen original para que coincida
img_resized = cv2.resize(img_rgb, (220, 220))

# Armá un mosaico 2 x 2.
fila_superior = cv2.hconcat([img_resized, roi_grande])
fila_inferior = cv2.hconcat([roi_grande, img_resized])
mosaico = cv2.vconcat([fila_superior, fila_inferior])

mostrar_imagenes(
    [roi, roi_grande, mosaico],
    ["ROI", "ROI redimensionada", "Mosaico"],
    figsize=(15, 4),
)
```

## Ejercicio 3: Bordes y umbralización

```python
img_texto = cv2.imread("texto.png", cv2.IMREAD_GRAYSCALE)

canny_a = cv2.Canny(img_texto, 50, 150)
canny_b = cv2.Canny(img_texto, 100, 200)

mostrar_imagenes(
    [img_texto, canny_a, canny_b],
    ["Texto original", "Canny A", "Canny B"],
    figsize=(15, 4),
)

img_monedas = cv2.imread("monedas.png", cv2.IMREAD_GRAYSCALE)

seg_a = segmentar_por_umbral(img_monedas, 80)
seg_b = segmentar_por_umbral(img_monedas, 150)

mostrar_imagenes(
    [img_monedas, seg_a, seg_b],
    ["Monedas originales", "Umbral A", "Umbral B"],
    figsize=(15, 4),
)
```

#### CUANDO CONVIENE MIRAR CANALES POR SEPARADO?
Mirar cada canal de color por separado es útil cuando:
* Querés destacar información específica de un color. Por ejemplo, la vegetación es más evidente en el canal verde; el cielo o el agua destacan en azul.
* Detección de objetos o segmentación basada en color. Si un objeto es rojo, mirar solo el canal rojo puede ayudar a separarlo del fondo.
* Corrección o análisis de color. Permite detectar sobreexposición en un canal, colores apagados, o problemas de balance.

#### QUE CAMBIA AL MODIFICAR LOS UMBRALES DE CANNY?
* Usamos dos umbrales (valores), umbral_min --> bordes débiles, umbral_max --> bordes fuertes.
1. Ambos umbralos bajos: muchos bordes, ruido?
2. Ambos umbrales altos: bordes mas mnotorios, mas marcados. Perdemos detalle también.
3. Moderado (min < max): lo mas equilibrado.

#### POR QUE UNA SEGMENTACION POR UMBRAL PUEDE FUNCIONAR BIEN EN UNA IMAGEN Y MAL EN OTRA?
La segmentación por umbral depende de la intensidad de píxels:
* Va a andar bien si el objeto y el fondo tienen contrastres claros.
* Va a andar mal si hay poca diferencia de intensidad.
